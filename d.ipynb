```python
def cols_cat(col,x):
    cat=''
    
    if df[col].describe()[3]<= x < df[col].describe()[4]:
        cat=1
    elif df[col].describe()[4]<= x < df[col].describe()[5]:
        cat=2
    elif df[col].describe()[5]<= x < df[col].describe()[6]:
        cat=3
    else: cat=4
    
    return cat
```

```python
df['cdepth_cat']=df['cdepth'].apply(lambda x: cols_cat('cdepth',x))

tst_df['cdepth_cat']=tst_df['cdepth'].apply(lambda x: cols_cat('cdepth',x))
```

```python
df.head(3)
```

```python
# cdepth그룹별 nrate / crate
grouped_nrate= df['nrate'].groupby(df['cdepth_cat'])
grouped_crate= df['crate'].groupby(df['cdepth_cat'])

print(grouped_nrate.mean())
print(grouped_crate.mean())
```

```python
def get_model_cv_prediction(model, X_data, y_target):
    neg_mse_scores = cross_val_score(model, X_data, y_target, scoring="neg_mean_squared_error", cv = 5)
    rmse_scores  = np.sqrt(-1 * neg_mse_scores)
    avg_rmse = np.mean(rmse_scores)
    print('##### ',model.__class__.__name__ , ' #####')
    print(' 5 교차 검증의 평균 RMSE : {0:.3f} '.format(avg_rmse))
```

```python
from bayes_opt import BayesianOptimization
```

```python
def get_score(y_test, pred):
    mse = mean_squared_error(y_test, pred)
    rmse = np.sqrt(mse)
    return rmse
```

```python
def scorer(model, X, y):
    pred = model.predict(X)
    return get_score(y, pred)
```

```python
bayesian_params = {
    'a':(0, 30)
}

def lasso_eval_train(a):
    lasso = Lasso(alpha=a)
    scores = cross_val_score(lasso, x_train, y_train, cv=50, scoring=scorer)
    
    return -np.mean(scores)
```

```python
lassoBO = BayesianOptimization(lasso_eval_train, bayesian_params)
lassoBO.maximize(init_points=10, n_iter=50)
```

```python

def lasso_eval_val(a):
    lasso = Lasso(alpha=a)
    scores = cross_val_score(lasso, x_val, y_val, cv=20, scoring=scorer)
    
    return -np.mean(scores)

lassoBO = BayesianOptimization(lasso_eval_val, bayesian_params)
lassoBO.maximize(init_points=10, n_iter=50)
```

```python
results = []
for result in lassoBO.res:
    results.append(result['target'])
max_ind = np.argmax(results)
lassoBO.res[max_ind]
```
