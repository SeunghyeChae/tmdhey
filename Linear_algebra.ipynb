### Summary of Projection

- The scalar projection of a vector u onto a vector v is q ⋅ u, where q is the unit vector in the direction of v. 

- The vector projection of u onto v is the scalar projection of u onto v times q, where q is the unit vector in the direction of v.

- The vector projection of u onto v is the best approximation of u in the direction of v, in the sense that the difference between u and its vector projection onto v is orthogonal to v.

- The work done by a force that is applied at an angle to the displacement vector can be computed by projecting the force vector onto the displacement vector, and then multiplying the magnitudes of the force and displacement vectors.

```python
# Pythagorean theorem.
# c^ = a^+b^ (Triangle Inequality)

# simpple python code for triangle inequality

import numpy as np

def size_of_vector(vector):
  return np.sqrt(np.sum(np.square(vector)))

a= np.array([3,0,4])
b= np.array([0,5,12])

res1= size_of_vector(a+b)
res2= size_of_vector(a) + size_of_vector(b)

print('|a+b| : ', res1)
print('|a|+|b| : ', res2)
```

```python
# changing basis (co-ordinate system)

# 딥러닝에서 많이 하는 것이 어떤 문제를 다른 고밀도 벡터로 표현하는 방법이 많은데
# 수학적으로는 하나의 coordinate system에서 다른 coordicate system으로 바꾸는 것과 
# 비슷하다고 한다. 대부분의 머신러닝 문제가 이것과 상당히 밀접한 관련이 있다고함
# 좌표계(coordinate system) - 숫자나 좌표를 이용해서 유클리디안 공간안에서 기하학적으로
# 문제를 표현하는 시스템

# computation을 빠르게 하려면 벡터들을 직교로 만들어주는게 좋은데 
# 기저변환(changing basis)을 통해서 직교 상태로 basis axis를 
# 만들어 놓고 새롭게 벡터를 구할 수도 있다
```

```python
# re 벡터에 대해서 새로운 basis(기저)로 표현하고자 할때 그 projection
import numpy as np
def sov(vector): #size of vector
  return np.sqrt(np.sum(np.square(vector)))

re= np.array([3,4])
b2= np.array([-2,4])

y= np.dot(re,b2)/np.square(sov(b2))

print(y)
```

```python
#size of vector
def sov(vector):
  return np.sqrt(np.sum(np.square(vector)))

# 기저 변환 함수
def change_basis(v, b1, b2):
  vb1 = np.dot(v, b1) / np.square(sov(b1)) #  vector projection 구하기
  vb2 = np.dot(v, b2) / np.square(sov(b2))
  # print(vb1, ',', vb2)
  return np.array([vb1, vb2])

v = np.array([10, -5])
b1 = np.array([3, 4])
b2 = np.array([4, -3])
vb = change_basis(v, b1, b2)

print(vb)
```
